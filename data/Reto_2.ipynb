```python
def clasificar(fila):
    if fila["pH"] > 8.5 and fila["Temp"] > 25:
        return "Critico"
    elif fila["OD_promedio"] < 5:
        return "Bajo Oxigeno"
    else:
        return "Normal"

datos["Categoria"] = datos.apply(clasificar, axis=1)
```

```python
datos["Categoria"].value_counts()
```

```python
datos.to_csv("reto2_resultado.csv", index=False)
```
