```python
import pandas as pd

datos = pd.read_csv('4_resultados_agua.csv')
datos.head()
```

```python
datos.info()
datos.describe()
```

```python
def clasificar_ph(valor):
    if valor < 6.5:
        return 'Ácido'
    elif valor <= 8.5:
        return 'Neutro'
    else:
        return 'Básico'

datos['categoria_ph'] = datos['pH'].apply(clasificar_ph)
datos.head()
```

```python
agua_acida = datos[datos['categoria_ph'] == 'Ácido']
agua_acida.head()
```

```python
datos.mean(numeric_only=True)
```

```python
datos['pH'].mean()
```

```python
datos.groupby('categoria_ph').mean(numeric_only=True)
```

```python
datos['calidad'] = datos['pH'] * 10  # ejemplo simple
```

```python
datos['ph_normalizado'] = (datos['pH'] - datos['pH'].mean()) / datos['pH'].std()
```

```python
resumen = datos.groupby('categoria_ph').agg({
    'pH': ['mean', 'min', 'max'],
})
resumen
```
